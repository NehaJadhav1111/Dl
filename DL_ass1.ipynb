```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
```

```python
Y = np.array([0,0,0,1])
```

```python
np.random.seed(1)
weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1

print("Inital Weights : ", weights)
print("Intial Bias :", bias )
```

```python
def step_function(X):
  return np.where( X >= 0,1,0)
```

```python

epochs = 10

for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)

        # Calculate error
        error = Y[i] - y_pred

        # Update rule
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Error: {error}")

print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```

```python
for i in range(len(X)):
  y_pred = step_function(np.dot(X[i], weights)+ bias)
  print(f"Input: {X[i]}, Output: {y_pred}")
```

```python
X1 = np.linspace(-0.5, 1.5, 100)
X2 = -(weights[0] * X1 + bias) / weights[1]

plt.pyplot.figure(figsize=(6,6))
plt.pyplot.plot(X1, X2, 'r-', label='Decision Boundary')

for i in range(len(X)):
    if Y[i] == 0:
        plt.pyplot.scatter(X[i][0], X[i][1], color='blue', marker='o', label='Class 0' if i==0 else "")
    else:
        plt.pyplot.scatter(X[i][0], X[i][1], color='green', marker='x', label='Class 1' if i==3 else "")

plt.pyplot.xlabel('X1')
plt.pyplot.ylabel('X2')
plt.pyplot.legend()
plt.pyplot.title('Perceptron Decision Boundary (AND Gate)')
plt.pyplot.grid(True)
plt.pyplot.show()
```
